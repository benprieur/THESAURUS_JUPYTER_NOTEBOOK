```python
import os
import pymysql
import pandas as pd

conn = pymysql.connect(
    host="localhost",
    user="root",
    passwd="root",
    db="THESAURUSDB")
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TREVISION",
    conn)
df.head(1)
# Nombre d'éditions de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TUSER",
    conn)
df.head(1)
# Nombre d'édit.eur.rice.s de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TTHESAURUS",
    conn)
df.head(1)
# Nombre de thésaurus de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT TUSER.userName,COUNT(*) AS NB FROM TREVISION, TUSER WHERE TUSER.id=TREVISION.idUser GROUP BY idUser Order by NB DESC",
    conn)
```

```python
df['NB'].median()
# Médiane des éditions par utilisat.eur.rice ayant édité au moins une fois un thésaurus en français sur fr.wiktionary
# Beaucoup à une édition
```

```python
df['NB'].mean()
# Moyenne des éditions = 15
```

```python
df.head(50)
# Top 50
```

```python
df.groupby(df.index // 250).sum()['NB'][0]
# Le nombre d'éditions des 250 premiers
```

```python
df.groupby(df.index // 250).sum()['NB'][0] / 8801 * 100
# Les 255 premiers représentent > 95 % des éditions
```

```python
df.groupby(df.index // 255).median()['NB'][0]
# Médiane du nombre d'éditions par utilisat.eur.rice sur les 255 premiers users
```

```python
df2 = pd.read_sql_query("SELECT TTHESAURUS.name,COUNT(*) AS NB FROM TREVISION, TTHESAURUS WHERE TTHESAURUS.id=TREVISION.idThesaurus GROUP BY idThesaurus Order by NB DESC", conn)
# On s'intéresse ici au nombre d'éditions par thésaurus
```

```python
df2['NB'].median()
# médiane du nombre d'éditions de chaque thésaurus
```

```python
df2['NB'].mean()
# moyenne du nombre d'éditions de chaque thésaurus
```

```python
df2.head(50)
# Top 50 des thésaurus les plus édités
```

```python
df2.groupby(df2.index // 250).sum()['NB'][0]
```

```python
df2.groupby(df2.index // 250).sum()['NB'][0] / 8801 * 100
# 250 thésaurus représentent environ 95 % des éditions
```

```python
df4 = pd.read_sql_query(
    "SELECT TTHESAURUS.name, TUSER.userName, TREVISION.instant FROM TREVISION, TTHESAURUS, TUSER WHERE TTHESAURUS.id=TREVISION.idThesaurus And TUSER.id=TREVISION.idUser",
    conn)
df4.head()
# Tableau technique pour analyser la création des thésaurus
```

```python
df4.groupby("name", as_index=False)["instant"].min()
```

```python
df5 = df4.iloc[df4.groupby("name", as_index=False)["instant"].idxmin()]
```

```python
values = df5['userName'].value_counts().keys().tolist()
counts = df5['userName'].value_counts().tolist()
```
