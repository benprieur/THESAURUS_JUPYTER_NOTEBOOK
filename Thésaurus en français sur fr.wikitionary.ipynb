```python
import os
import pymysql
import pandas as pd

conn = pymysql.connect(
    host="localhost",
    user="root",
    passwd="root",
    db="THESAURUSDB")
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TREVISION",
    conn)
df.head(1)
# Nombre d'éditions de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TUSER",
    conn)
df.head(1)
# Nombre d'édit.eur.rice.s de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT COUNT(*) FROM TTHESAURUS",
    conn)
df.head(1)
# Nombre de thésaurus de thésaurus en français sur fr.wikt
```

```python
df = pd.read_sql_query(
    "SELECT TUSER.userName,COUNT(*) AS NB FROM TREVISION, TUSER WHERE TUSER.id=TREVISION.idUser GROUP BY idUser Order by NB DESC",
    conn)
```

```python
df['NB'].median()
# Médiane des éditions par utilisat.eur.rice ayant édité au moins une fois un thésaurus en français sur fr.wiktionary
# Beaucoup à une édition
```

```python
df['NB'].mean()
# Moyenne des éditions = 15
```

```python
df.head(50)
# Top 50
```

```python
df.groupby(df.index // 250).sum()['NB'][0]
# Le nombre d'éditions des 250 premiers
```

```python
df.groupby(df.index // 250).sum()['NB'][0] / 8801 * 100
# Les 255 premiers représentent > 95 % des éditions
```

```python
df.groupby(df.index // 255).median()['NB'][0]
# Médiane du nombre d'éditions par utilisat.eur.rice sur les 255 premiers users
```
